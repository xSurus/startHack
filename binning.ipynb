# Binning

Tip on creating a new features;
We can categorise the elevation variables through binning.
For example: Elevations from 0 - 100 can be categorised as steep, 100 - 120 very steep and maybe elevations with more than 120 can categorised as extremely steep.
Experiment with different bins and choose the ones with the highest performance.

### Elevation

in second argument of pd.cut you can configure the bins, and label is the name of respective bin
so here 0-99 is bin 0, 100-199 is bin 1, 200-999 is bin 2, else it's NaN

```python
import pandas as pd
import numpy as np
from dfcols import all_square_cols

train = pd.read_csv('./data/Train.csv')
test = pd.read_csv('./data/Test.csv')
sample_submission = pd.read_csv('./data/SampleSubmission.csv')

elevationCols = all_square_cols("elevation")

elevationCols = train[elevationCols]

for col in elevationCols:
    curCol = elevationCols[col]
    elevationCols[col] = pd.cut(curCol, [0,100,200,1000], labels=[0,1,2]) # here

train_binning = train.replace(elevationCols)

train_binning.describe(include = 'all')
```
